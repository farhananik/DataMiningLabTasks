```python
import pandas as pd
from scipy.spatial import distance
from scipy.spatial.distance import cityblock, euclidean, minkowski
```

```python
df = pd.read_csv('dataset_review.csv')
df
```

```python
df = df.replace('-', float('NaN'))
```

```python
df
```

```python
df1
```
