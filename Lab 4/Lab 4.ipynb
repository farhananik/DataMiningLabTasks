```python
dataset = [['milk','eggs','bread'],
          ['milk','bread'],
          ['eggs','sugar'],
          ['coffee','sugar'],
          ['onion','coffee','carrot','icecream']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
encoder = TransactionEncoder().fit(dataset)
```

```python
onehot = encoder.transform(dataset)
```

```python
onehot1 = pd.DataFrame(onehot,columns = encoder.columns_)
```

```python
onehot1
```

```python
import scipy.spatial
```

```python
jaccard = scipy.spatial.distance.cdist(onehot1, onehot, metric = 'jaccard')
```

```python
jaccard
```

```python
jaccarddis = pd.DataFrame(jaccard,columns=onehot1.index.values,index=onehot1.index.values)
```

```python
jaccarddis
```

```python
from scipy.spatial import distance
```

```python
man_Dist0 = distance.cityblock(jaccarddis[0],1)
man_Dist1 = distance.cityblock(jaccarddis[1],1)
man_Dist2 = distance.cityblock(jaccarddis[2],1)
man_Dist3 = distance.cityblock(jaccarddis[3],1)
man_Dist4 = distance.cityblock(jaccarddis[4],1)
```

```python
print(man_Dist0)
print(man_Dist1)
print(man_Dist2)
print(man_Dist3)
print(man_Dist4)
```
