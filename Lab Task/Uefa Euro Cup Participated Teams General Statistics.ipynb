```python
import pandas as pd
```

```python
df = pd.read_csv('Uefa Euro Cup Participated Teams General Statistics.csv')
```

```python
df
```

```python
print(df.columns)
```

```python
from scipy.spatial import distance
from scipy.spatial.distance import cityblock, euclidean
from sklearn.metrics.pairwise import cosine_similarity
```

```python
man_Dist = distance.cityblock(df['Goal_For'].astype(float), df['Goal_Against'].astype(float))
euc_Dist = distance.euclidean(df['Goal_For'].astype(float), df['Goal_Against'].astype(float))
```

```python
print('\t\t', 'Manhattan', '\t', 'Euclidean')
print('GF - GA', '\t' ,man_Dist, '\t\t' ,euc_Dist)
```

```python
print(cosine_similarity([df.Goal_For], [df.Goal_Against]))
```
