```python
import numpy as np
```

```python
print(np.cov(df2.NA_Sales, df2.Global_Sales))
```

```python
def pearson(x,y):
    correlation_matrix=np.corrcoef(x,y)
    return correlation_matrix[0,1]
```

```python
print(pearson(df2.NA_Sales, df2.Global_Sales))
```
